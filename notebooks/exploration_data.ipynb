# Exploration des données - Fraude Assurance

Notebook simple pour analyser le dataset `insurance_claims.csv`.


```python
import pandas as pd
from pathlib import Path

data_path = Path('../data/insurance_claims.csv')
df = pd.read_csv(data_path)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['fraud_reported'].value_counts(normalize=False)
```

```python
df['fraud_reported'].value_counts(normalize=True) * 100
```

```python
df.replace('?', pd.NA).isna().sum().sort_values(ascending=False).head(15)
```

```python
numeric_cols = df.select_dtypes(include='number').columns.tolist()
categorical_cols = [c for c in df.columns if c not in numeric_cols]
print('Variables numériques:', len(numeric_cols))
print(numeric_cols)
print('Variables catégorielles:', len(categorical_cols))
print(categorical_cols)
```

## Interprétation métier

Le dataset contient des informations sur l'assuré, la police d'assurance, le sinistre, les montants réclamés et le véhicule. La cible `fraud_reported` permet d'entraîner un modèle supervisé pour estimer le risque de fraude.
